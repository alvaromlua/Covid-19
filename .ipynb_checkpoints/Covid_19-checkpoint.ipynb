```python
selected_df = df[df.finding=="Pneumonia/Viral/COVID-19"]
```

```python
selected_df['view'].value_counts(dropna=False)
```

```python
selected_df = selected_df[(selected_df.view == "AP") | (selected_df.view == "PA")]
selected_df.head(2)
```

```python
images = selected_df.filename.values.tolist()
```

```python
os.makedirs('./COVID19-DATASET/train/covid19')
os.makedirs('./COVID19-DATASET/train/normal')
```

```python
COVID_PATH = './COVID19-DATASET/train/covid19'
NORMAL_PATH = './COVID19-DATASET/train/normal'
```

```python
for image in images:
    shutil.copy(os.path.join('./covid-chestxray-dataset/images', image), os.path.join(COVID_PATH, image))
```

```python
for image in os.listdir('./COVID19-DATASET/train/covid19')[:300]:
    shutil.copy(os.path.join('./COVID19-DATASET/train/covid19', image), os.path.join(NORMAL_PATH, image))
```

```python
DATA_PATH = './COVID19-DATASET/train'
```

```python
class_names = os.listdir(DATA_PATH)
image_count = {}
for i in class_names:
    image_count[i] = len(os.listdir(os.path.join(DATA_PATH,i)))

#Plotting Distribution of Each Classes
fig1, ax1 = plt.subplots()
ax1.pie(image_count.values(),
        labels = image_count.keys(),
        shadow=True,
        autopct = '%1.1f%%',
        startangle=90)
plt.show()
```

```python
fig = plt.figure(figsize=(16,5))
fig.suptitle("COVID19 Positive", size=22)
img_paths = os.listdir(COVID_PATH)
shuffle(img_paths)

for i,image in enumerate(img_paths[:4]):
    img = cv2.imread(os.path.join(COVID_PATH, image))
    plt.subplot(1,4, i+1, frameon=False)
    plt.imshow(img)
fig.show()
```

```python
fig = plt.figure(figsize=(16,5))
fig.suptitle("COVID19 Negative - Healthy", size=22)
img_paths = os.listdir(NORMAL_PATH)
shuffle(img_paths)

for i,image in enumerate(img_paths[:4]):
    img = cv2.imread(os.path.join(NORMAL_PATH, image))
    plt.subplot(1,4, i+1, frameon=False)
    plt.imshow(img)
fig.show()
```

## **Transforma**
La transformación o aumento de datos es la síntesis de nuevos datos utilizando los disponibles con algunas pequeñas manipulaciones y procesamiento de imágenes. El aumento ayudará a generalizar nuestro modelo y evitará el sobreajuste de los datos de entrenamiento. Dado que tenemos relativamente poca cantidad de datos para entrenamiento y validación, sintetizaremos algunos datos adicionales a través de Transformaciones de imágenes.

torchvision de PyTorch proporciona varias herramientas que podemos utilizar para realizar diversas tareas de visión por computadora con facilidad de uso.

Algunas de las herramientas más utilizadas incluyen,

transforma: aumento de datos de imágenes
conjuntos de datos: carga y manejo de conjuntos de datos
modelos: modelos SOTA predefinidos de aprendizaje profundo
Estas herramientas nos serán muy útiles para que podamos concentrarnos en optimizar mejor nuestros resultados.
